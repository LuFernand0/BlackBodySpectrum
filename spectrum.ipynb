```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
```

```python
# Constantes físicas
h = 6.62607015e-34   # constante de Planck
c = 3e8              # velocidade da luz
k = 1.380649e-23     # constante de Boltzmann
```

```python
# Lei de Planck
def planck(wavelength, T):
    a = 2.0*h*c**2
    b = h*c/(wavelength*k*T)
    intensity = a / (wavelength**5 * (np.exp(b) - 1.0))
    return intensity
```

```python
def plot_blackbody(T):
    wavelengths = np.linspace(1e-9, 3e-6, 500)  # 1 nm a 3000 nm
    intensity = planck(wavelengths, T)
    
    plt.figure(figsize=(8,5))
    plt.plot(wavelengths*1e9, intensity)
    plt.xlabel("Comprimento de onda (nm)")
    plt.ylabel("Intensidade")
    plt.title(f"Espectro de Corpo Negro — T = {T} K")
    plt.grid(True)
    plt.show()
```

```python
interact(plot_blackbody, T=FloatSlider(min=500, max=10000, step=100, value=3000));
```
